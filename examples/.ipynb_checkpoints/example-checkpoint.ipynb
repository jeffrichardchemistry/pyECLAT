```python
import pandas as pd
import numpy as np
```

```python
path_csvfile = 'pyECLAT/data/base2.csv' #path of csv file
df_base2 = pd.read_csv(path_csvfile, header=None) #create dataframe. Its need set up header=None
df_base2
```

```python
from pyECLAT import ECLAT
```

```python
eclat_class = ECLAT(data=df_base2, verbose=True)
```

```python
#Cheking the binary dataframe generate
#Check if a dataframe like this has been generated. The product names are now in the column name (no repetition)
eclat_class.df_bin
```

```python
get_ECLAT_indexes, get_ECLAT_supports = eclat_class.fit(min_support=0.04,
                                                           min_combination=1,
                                                           max_combination=4,
                                                           separator=' & ',
                                                           verbose=True)
```

```python
#Supports values of products 
get_ECLAT_supports
```

```python
#Index of the occurrence of products together
#This index refers to the original dataframe
get_ECLAT_indexes
```

```python
#This function will search for all possible combinations that have support > 0.02 (in this case)
get_ECLAT_indexes_all, get_ECLAT_supports_all = eclat_class.fit_all(min_support=0.02,
                                                           separator=' & ',
                                                           verbose=True)
```
